## 1.1激活函数
**sigmoid 函数**
$$\sigma = \frac{1}{1+e^{z}} $$

**sigmoid函数求导的特点**
$$ \frac{dy}{dx} = \sigma(x)(1-\sigma(x)) $$



```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    """
    y = 1/(1+e^-x)

    """
    y = 1/(1+np.exp(-z))
    return y 

def sigmoid_derivative(x):
    return sigmoid(x)*(1-sigmoid(x))

x = np.linspace(-10,10,10000)
y = sigmoid(x)
plt.plot(x,y)
plt.title('sigmoid function')
plt.grid(True)
```
